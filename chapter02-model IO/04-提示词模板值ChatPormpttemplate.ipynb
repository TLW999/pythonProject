##2.调用提示词模板的几种方法：invoke() \ format() \ format_messages() \ format_prompt()

invoke()：传入的是dict ；返回值：ChatPromptValue

format() : 传入的是：变量的值    ；返回值：str

format_messages(): 传入的是：变量的值    ；返回值：消息构成的list

format_prompt(): 传入的是：变量的值   ；返回值：ChatPromptValue


###举例1：format()

```python
chat_prompt_template=ChatPromptTemplate.from_messages([
    ("system","你是一个Ai助手，你的名字叫{name}"),
    ("human","我的问题是{question}"),
])

response = chat_prompt_template.format(name="小智",question="1+1*4=?")

print(response)
```

如何实现ChatPromptValue与list[message]、字符串之间的转换

```python
chat_prompt_template=ChatPromptTemplate.from_messages([
    ("system","你是一个Ai助手，你的名字叫{name}"),
    ("human","我的问题是{question}"),
])

response = chat_prompt_template.format_prompt(name="小智",question="1+1*4=?")


#将ChatPromptValue类型转换为消息构成的list
response_messages = response.to_messages()
print(type(response)) #<class 'list'>


# 将ChatPromptValue类型转换为消息构成为字符串类型
response_to_string=response.to_string()
print(type(response_to_string))
```

集合LLM

```python
from langchain_core.messages import SystemMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
import os
import dotenv
from langchain_core.prompts import SystemMessagePromptTemplate,HumanMessagePromptTemplate
dotenv.load_dotenv()
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
os.environ["OPENAI_BASE_URL"]=os.getenv("OPENAI_BASE_URL")

#创建大模型实例
chat_model = ChatOpenAI(
    model_name="gpt-4o-mini",
    temperature = 0.7,
)

#通过chat提示词模板，创建提示词
system_template = "你是一个专家{role}"
system_message_prompt=SystemMessagePromptTemplate.from_templates=(system_template)

human_template = "给我解释{content},用浅显易懂的语言"
human_message_prompt = HumanMessagePromptTemplate.from_template(human_template)

chat_prompt_template=ChatPromptTemplate.from_messages(
    [
        system_message_prompt,human_message_prompt
    ]
)

prompt=chat_prompt_template.invoke({"role":"物理学家","content":"相对论"})


#通过大模型调用提示词，得到响应数据
response = chat_model.invoke(prompt)

print(response.content)
```

5.插入消息列表：MessagePlaceholder
使用场景：当ChatPromptTemplate模板中的消息类型和个数不确定的时候，我们就可以使用MessagePlaceholder

```python
from langchain_core.messages import HumanMessage
from langchain_core. prompts. chat import MessagesPlaceholder

chat_prompt_template=ChatPromptTemplate([
        ("system","你是一个Ai助手，你的名字叫{name}"),
        MessagesPlaceholder(variable_name="msgs")
    ])

response = chat_prompt_template.invoke({
    "name":"小智",
    "msgs":[HumanMessage(content="我的问题是：1+3*4=？")]
})

print(response)
```

```python
from langchain_core.messages import HumanMessage,AIMessage
from langchain_core. prompts. chat import MessagesPlaceholder

chat_prompt_template=ChatPromptTemplate([
        ("system","你是一个Ai助手，你的名字叫{name}"),
        MessagesPlaceholder(variable_name="msgs")
    ])

response = chat_prompt_template.invoke({
    "name":"小智",
    "msgs":[HumanMessage(content="我的问题是：1+3*4=？"),AIMessage(content="3+9*8=?")]
})

print(response)
```

存储对话历史记录

```python
prompt = ChatPromptTemplate.from_messages([
    ("system","you are helpful assistant."),
    MessagesPlaceholder("history"),
    ("human","{question}")
])

prompt_value = prompt.format_prompt(
    history=[HumanMessage(content="1+3*4=？"),AIMessage(content="3+9*8=?")],
    question = "我刚才问题是什么"
)

response = chat_model.invoke(prompt_value)
print(response.content)
```
