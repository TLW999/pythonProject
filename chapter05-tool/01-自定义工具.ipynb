# 1. 使用@tool装饰器定义工具

```python
from langchain_core.tools import tool



@tool
def add_numbers(a: int, b: int):
    """计算两个整数和"""
    return a + b


print(f"name = {add_numbers.name}")  #默认是函数的名称
print(f"args_schema = {add_numbers.args_schema}")
print(f"description = {add_numbers.description}")  #默认是函数的说明信息
print(f"return_direct = {add_numbers.return_direct}")  #默认是false
```

```python
from langchain_core.tools import tool


@tool(name_or_callable="add_two_numbers", description="caculate numbers", return_direct=True)
def add_numbers(a: int, b: int):
    """计算两个整数和"""
    return a + b


print(f"name = {add_numbers.name}")  #默认是函数的名称
print(f"args_schema = {add_numbers.args_schema}")
print(f"description = {add_numbers.description}")  #默认是函数的说明信息
print(f"return_direct = {add_numbers.return_direct}")  #默认是false
```

```python
#调用工具
res = add_numbers.invoke({"a": 10, "b": 20})
print(res)
```

```python
from pydantic import Field
from openai import BaseModel
from langchain_core.tools import tool


class FieldInfo(BaseModel):
    a: int = Field(description="第1个参数")
    b: int = Field(description="第2个参数")


@tool(name_or_callable="add_two_numbers", description="caculate numbers", return_direct=True,args_schema=FieldInfo)
def add_numbers(a: int, b: int):
    """计算两个整数和"""
    return a + b


print(f"name = {add_numbers.name}")  #默认是函数的名称
print(f"args = {add_numbers.args}")
print(f"description = {add_numbers.description}")  #默认是函数的说明信息
print(f"return_direct = {add_numbers.return_direct}")  #默认是false
```

# 2.StructuredTool的from_function()的使用

举例1

```python
from langchain_core.tools.structured import StructuredTool

class FieldInfo(BaseModel):
    query: str = Field(description="要搜索的关键词")


#1.声明一个函数
def search_google(query: str):
    return "最后查询结果"

#定义一个工具
search01=StructuredTool.from_function(
    func=search_google,
    name="Search",
    description="查询google搜索引擎，并讲结果返回",
    return_direct=True,
    args_schema=FieldInfo,
)

print(f"name = {search01.name}")
print(f"args = {search01.args}")
print(f"description = {search01.description}")
print(f"return_direct = {search01.return_direct}")

search01.invoke({"query":"中美现状"})
```
